1. Berapa rata-rata tip yang diberikan penumpang selama tahun 2022?

```python
df_1 = client.query('''

SELECT
  FORMAT('$%.2f', AVG(tip_amount)) AS tip_passenger
FROM `bigquery-public-data.new_york_taxi_trips.tlc_yellow_trips_2022`
LIMIT 5000

''').to_dataframe()

df_1
```

2. Pada bulan berapa rata-rata total_amount paling tinggi sepanjang tahun 2022?

```python
df_1 = client.query('''

SELECT
  EXTRACT(MONTH FROM dropoff_datetime) as month_time,
  FORMAT("$%.2f", COUNT(*) / COUNT(DISTINCT EXTRACT(YEAR FROM dropoff_datetime))) AS average_amount
FROM `bigquery-public-data.new_york_taxi_trips.tlc_yellow_trips_2022`
GROUP BY month_time
ORDER BY average_amount DESC
LIMIT 1

''').to_dataframe()

df_1
```

3. Tampilkan 5 vendor (berdasarkan vendor_id) dengan rata-rata tip_amount tertinggi.

```python
df_1 = client.query('''

SELECT
  vendor_id,
  FORMAT("$%.2f", AVG(tip_amount)) AS average_tip
FROM `bigquery-public-data.new_york_taxi_trips.tlc_yellow_trips_2022`
GROUP BY vendor_id
ORDER BY average_tip DESC
LIMIT 5

''').to_dataframe()

df_1
```

4. Berapa banyak trip yang terjadi pada bulan Desember 2022?

```python
df_1 = client.query('''

SELECT
  COUNT(*) AS total_trip
FROM `bigquery-public-data.new_york_taxi_trips.tlc_yellow_trips_2022`
WHERE EXTRACT(MONTH FROM dropoff_datetime) = 12

''').to_dataframe()

df_1
```

5. Tampilkan 10 pickup_location_id yang paling sering digunakan.

```python
df_1 = client.query('''

SELECT
  pickup_location_id,
  COUNT(*) AS total_trip
FROM `bigquery-public-data.new_york_taxi_trips.tlc_yellow_trips_2022`
GROUP BY pickup_location_id
ORDER BY total_trip DESC
LIMIT 10

''').to_dataframe()

df_1
```

6. Tampilkan hari (Seninâ€“Minggu) mana yang paling sering terjadi trip.

```python
df_1 = client.query('''

SELECT
  FORMAT_DATE("%A", dropoff_datetime) AS day_of_week,
  COUNT(*) AS total_trip
FROM `bigquery-public-data.new_york_taxi_trips.tlc_yellow_trips_2022`
GROUP BY day_of_week
ORDER BY total_trip DESC
LIMIT 5

''').to_dataframe()

df_1
```

7. Di jam berapa rata-rata trip paling sering terjadi? (Gunakan jam dari pickup_datetime)

```python
df_1 = client.query('''

SELECT
  EXTRACT(HOUR FROM pickup_datetime) AS hour_time,
  COUNT(*) AS total_trip
FROM `bigquery-public-data.new_york_taxi_trips.tlc_yellow_trips_2022`
GROUP BY hour_time
ORDER BY total_trip DESC
LIMIT 1

''').to_dataframe()

df_1
```
